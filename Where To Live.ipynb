# Data Preparation

### Add Explicit State Abbreviation to Demo_df
Start by adding a state abbreviation field to the demo_df dataset by matching on the numeric state designator shared by demo_df and demo_state_df

```python
#Use demo_state_df STATE_FIPS column to determine the state in demo_df

#Match on the state numeric code
for name in range(len(demo_df['STATE'])):
    for stateCode in range(len(demo_state_df['STATE_FIPS'])):
        if(demo_df['STATE'][name] == demo_state_df['STATE_FIPS'][stateCode]):
            demo_df['STATE'][name] = demo_state_df['STATE_ABBR'][stateCode]
```

### Split County Name and State Abbreviation

The county name and the state abbreviation are listed together as a string in the crime_df. Mutate field from a string to a list of strings by splitting on the comma. This will help for matching on the state later.

```python
#Split string on comma and turn into a list. Needed for matching later.
for name in range(len(crime_df['county_name'])):
    crime_df['county_name'][name] = crime_df['county_name'][name].split(",")
    crime_df['county_name'][name][1] = crime_df['county_name'][name][1].strip()
```

### Add a dummy join column
Add a join column to keep record of rows that refer to the same county in the crime_df and the demo_df

```python
# Add columns for matching
crime_df.insert(0,"JoinColumn", [np.nan] * crime_df.shape[0], True )
demo_df.insert(0,"JoinColumn", [np.nan] * demo_df.shape[0], True )
```

### Match crime_df and demo_df
Match the crime_df and the demo_df on both county and state, as states can have counties with the same name.

```python
# Match based on the county name and the state
numEqual = 0
for x in range(len(crime_df['county_name'])):
    for y in range(len(demo_df['CNTYNAME'])):
        if crime_df['county_name'][x][0] == demo_df['CNTYNAME'][y]:
            if(crime_df['county_name'][x][1] == demo_df['STATE'][y]):
                numEqual += 1
                crime_df['JoinColumn'][x] = numEqual
                demo_df['JoinColumn'][y] = numEqual
```

#### Matching result

The prints below show that we have 3135 matches and had to drop a little under 100 counties from the demo_df dataset. 

##### Note on data dropping
The counties that did not have a match have to be dropped because either crime or demographic data is missing. These counties just are not in both datasets.

One county was dropped from the crime_df dataset but 85 were dropped from demo_df. This means we were missing crime data for 85 counties.

Both the demo_df and crime_df have a unique county on each row, but demo_df has 84 more counties that crime_df, consequently, not all counties in demo_df can be mapped to crime in crime_df

```python
#Show the number of matches                
print(numEqual)
print(crime_df.shape)
print(demo_df.shape)

#Drop Columns that did not have a match
crime_df_dropped = crime_df.dropna(subset=['JoinColumn'], how = 'any')
demo_df_dropped = demo_df.dropna(subset=['JoinColumn'], how = 'any')

#Show the number of matches in each data frame
#Note that they are the same size, there were not any duplicate matches
print(crime_df_dropped.shape)
print(demo_df_dropped.shape)
```

### Aggregating Data and Imputation

Next, join the crime_df and demo_df datasets based on the matching column.

Impute values for the cities:
1. Columbia, Howard County
2. Odenton, Anne Arundel County
3. Glen Burnie, Anne Arundel County
4. Hanover, Anne Arundel County

Then drop the identification based fields.

#### On Imputation Methodology

I could not find a dataset for the cities that consisted of the same fields found in the demographic demo_df dataset. So I extract the rows for the counties the cites are located in then reduce the numeric fields by the difference in population. For example, if Howard County had 1000 Asian females and Columbia had 1/4 of the population of Howard County, then the value of the field for Asian females would be 250.


It basically assumes that the counties have a uniform demographic distribution. While that is is not a good assumption, the result is for personal estimation only, so I accept the inaccuracies associated with the assumption.

```python

#Columns that are not to be used in the modeling and will be dropped.
dropList = ['JoinColumn','county_name', 'EDITION', 'PART', 'IDNO', 'INDEX', 'FIPS_ST', 'FIPS_CTY','OBJECTID', 'STATE', 'STCN', 'CNTYNAME']

#Join on the matches
joinedDF = crime_df_dropped.merge(demo_df_dropped, on='JoinColumn')

#Subset by county based on potential living city locations
columbia = joinedDF[joinedDF['STATE'] =='MD'].copy()
columbia = columbia[columbia['CNTYNAME'] == "Howard County"]

odenton = joinedDF[joinedDF['STATE'] =='MD'].copy()
odenton = odenton[odenton['CNTYNAME'] == "Anne Arundel County"]

glen_burnie = joinedDF[joinedDF['STATE'] =='MD'].copy()
glen_burnie = glen_burnie[glen_burnie['CNTYNAME'] == "Anne Arundel County"]

hanover = joinedDF[joinedDF['STATE'] =='MD'].copy()
hanover = hanover[hanover['CNTYNAME'] == "Anne Arundel County"]

#Drop unneccessary columns
columbia = columbia.drop(dropList, axis =1)
odenton = odenton.drop(dropList, axis =1)
glen_burnie = glen_burnie.drop(dropList, axis =1)
hanover = hanover.drop(dropList, axis =1)
joinedDF = joinedDF.drop(dropList, axis =1)

# Population in 2019
# Used for imputation
howardCountyPop = 325690
columbiaPop = 103991
anneArundelPop = 579234
odentonPop = 41846
glen_burniePop = 69872
hanoverPop = 12952

#Impute the numeric fields based on the population proportion
def imputeSubset(df, portion):
    """
    Description: This function iterates through the columns of a dataframe
    and finds columns that are numeric. It then reduces or increases the values 
    of that column by the portion provided.

    Arguments:
        df: - the dataframe that is to be mutated 
        portion: - the amount to increase or decrease each value by.

    Returns:
        None: The passed dataframe is mutated.
    """
    for column in df.columns:
        if np.issubdtype(df.dtypes[column], np.number):
                df[column] = df[column] * portion

                
#Impute values for the cities from their county
imputeSubset(columbia, columbiaPop / howardCountyPop)
imputeSubset(odenton, odentonPop / anneArundelPop)
imputeSubset(glen_burnie, glen_burniePop / anneArundelPop)
imputeSubset(hanover, hanoverPop / anneArundelPop)
```

# Data Modeling

The function defined below is used to create linear regression models of:

1. Rape (RAPE)
2. Arson (ARSON)
3. Motor Vehicle Theft (MVTHEFT)
4. Larceny (LARCENY)
5. Burglary (BURGLRY)
6. Robbery (Robbery)
7. Aggravated Assult (AGASSLT)

for the cities: 

1. Columbia, Howard County
2. Odenton, Anne Arundel County
3. Glen Burnie, Anne Arundel County
4. Hanover, Anne Arundel County

#### Be sure to note the comment on imputation for the cities from the county data set

```python
def model(parameter, joinedDF):
    """
    Description: The function performs a linear regression on the passed data frame
    to predict the passed parameter. Note: A lot of the methodoloy here came from what
    I learned in the Udacity Data Science track.
    
    Arguments:
        parameter: The parameter to predict
        joinedDF: A cleaned dataframe that should contain the parameter of interest.
        
    Returns:
        List: A list containing the predictions for the cities of interest, the coeffiecents from
              the model, and a dataframe comparing the actual and predicted values.
    """
    X = joinedDF.drop([parameter], axis = 1)
    y = joinedDF[parameter]
    #Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state=101011)
    # train
    lm_model = LinearRegression(normalize=True) # Instantiate
    lm_model.fit(X_train, y_train) #Fit
    # predict
    y_test_preds = lm_model.predict(X_test)
    y_train_preds = lm_model.predict(X_train)
    #score
    test_score = r2_score(y_test, y_test_preds)
    train_score = r2_score(y_train, y_train_preds)
    print("The rsquared on the training data was {}.  The rsquared on the test data was {}.".format(train_score, test_score))
    
    #For review of the model's performance
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    coefs_df['abs_coefs'] = np.abs(lm_model.coef_)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    compareDF = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_preds})
    
    #Subset the city data based on the passed parameter
    columbiaDropped = columbia.drop([parameter], axis = 1)
    odentonDropped =odenton.drop([parameter], axis = 1)
    glen_burnieDropped = glen_burnie.drop([parameter], axis = 1)
    hanoverDropped = hanover.drop([parameter], axis = 1)
    
    #Attempt to predict the parameter using the model
    columbiaVal = lm_model.predict(columbiaDropped)[0] 
    odentonVal = lm_model.predict(odentonDropped)[0] 
    glen_burnieVal = lm_model.predict(glen_burnieDropped)[0] 
    hanoverVal = lm_model.predict(hanoverDropped)[0] 
    
    return [[columbiaVal, odentonVal,glen_burnieVal,hanoverVal], coefs_df, compareDF] 
```
